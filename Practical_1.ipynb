```python
# Looping through a list
fruits = ["apple", "banana", "cherry"]
for fruit in fruits:
    print(fruit)

# Looping through a range of numbers
for i in range(5):  # Iterates from 0 to 4
    print(i)
```

```python
# Counting down from 5
count = 5
while count > 0:
    print(count)
    count -= 1  # Decrement the counter
```

```python
# Logical AND
a = 5
b = 10
if a > 0 and b > 5:
    print("Both conditions are True")

# Logical OR
if a > 0 or b < 5:
    print("At least one condition is True")

# Logical NOT
if not a == 10:
    print("a is not equal to 10")
```

```python
# Check for even and odd numbers in a range
for i in range(1, 11):  # Numbers from 1 to 10
    if i % 2 == 0:
        print(f"{i} is even")
    else:
        print(f"{i} is odd")
```

```python
# Multiplication table
for i in range(1, 6):  # Rows
    for j in range(1, 11):  # Columns
        print(i * j, end=" ")
    print()  # Newline after each row
```

```python
# Using break
for i in range(1, 11):
    if i == 5:
        break  # Exit the loop
    print(i)

# Using continue
for i in range(1, 11):
    if i % 2 == 0:
        continue  # Skip the rest of the code for this iteration
    print(i)

# Else in loops
for i in range(5):
    print(i)
else:
    print("Loop completed without break")
```

# Importing Matplot Liabrary

```python
import matplotlib as matplotlib
import matplotlib.pyplot as plt
# Creating a 1D array for the x-axis
x = np.linspace(0, 10, 100)  # 100 points from 0 to 10

# Creating two y-axis datasets
y1 = np.sin(x)       # Sine wave
y2 = np.cos(x)       # Cosine wave
```

```python
# Creating a line plot
plt.figure(figsize=(10, 6))  # Optional: specify figure size

plt.plot(x, y1, label='Sine Wave', color='blue', linestyle='-', linewidth=2)
plt.plot(x, y2, label='Cosine Wave', color='red', linestyle='--', linewidth=2)

# Adding titles and labels
plt.title('Sine and Cosine Waves')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Adding a legend
plt.legend()

# Display the plot
plt.show()
```

```python
# Creating a scatter plot
plt.figure(figsize=(10, 6))

plt.scatter(x, y1, label='Sine Points', color='green', marker='o', alpha=0.7)
plt.scatter(x, y2, label='Cosine Points', color='purple', marker='x', alpha=0.7)

# Adding titles and labels
plt.title('Sine and Cosine Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Adding a legend
plt.legend()

# Display the plot
plt.show()
```

```python
# Creating sample categorical data
categories = ['A', 'B', 'C', 'D', 'E']
values1 = np.array([5, 7, 3, 8, 4])
values2 = np.array([6, 2, 5, 7, 3])

x_pos = np.arange(len(categories))  # Positions of the bars on the x-axis
width = 0.35  # Width of the bars

# Creating a bar chart
fig, ax = plt.subplots(figsize=(10, 6))

# Bars for the first dataset
bars1 = ax.bar(x_pos - width/2, values1, width, label='Dataset 1', color='skyblue')

# Bars for the second dataset
bars2 = ax.bar(x_pos + width/2, values2, width, label='Dataset 2', color='salmon')

# Adding titles and labels
ax.set_title('Bar Chart Comparison')
ax.set_xlabel('Categories')
ax.set_ylabel('Values')
ax.set_xticks(x_pos)
ax.set_xticklabels(categories)

# Adding a legend
ax.legend()

# Display the plot
plt.show()
```

```python
plt.figure(figsize=(10, 6))

plt.plot(x, y1, color='magenta', linestyle='-.', linewidth=2, label='Sine')
plt.plot(x, y2, color='cyan', linestyle=':', linewidth=2, label='Cosine')

plt.title('Customized Sine and Cosine Waves')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.xlim(0, 10)
plt.ylim(-1.5, 1.5)

plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.legend()

# Adding an annotation
plt.annotate('Sine Peak', xy=(np.pi/2, 1), xytext=(np.pi/2 + 1, 1.2),
             arrowprops=dict(facecolor='black', shrink=0.05))

plt.show()
```
